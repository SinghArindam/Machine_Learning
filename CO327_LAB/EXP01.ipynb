```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
array = np.array([1,2,3,4,5])
print("OUTPUT:\n\n")
print("Numpy Array :",array)
data={'Name':['John','Anna','Peter','Linda'],
     'Age':[28,24,35,32]}
df = pd.DataFrame(data)
print("Pandas DataFrame:\n",df)
x = np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
